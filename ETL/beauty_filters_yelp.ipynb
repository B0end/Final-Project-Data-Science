# Combining gmaps_metadata and yelp_business

## gmaps_metadata

```python
gmap_metadata_path = "gs://bucket123321/gmap_clean_data/metadata/filter/part-00000-a715b93b-bf1c-4057-abf4-6b3676e71cfc-c000.json"
gmap_metadata = pd.read_json(gmap_metadata_path, lines=True)
gmap_metadata.head(3)
```

```python
# Remove the specified columns
columns_to_drop = ['address', 'category', 'relative_results', 'url']
gmap_metadata = gmap_metadata.drop(columns=columns_to_drop)

# Rename the columns
column_mapping = {
    'gmap_id': 'id',
    'first_word': 'category',
    'State': 'state'
}
gmap_metadata = gmap_metadata.rename(columns=column_mapping)

gmap_metadata.head(3)
```

```python
gmap_metadata.shape
```

## yelp_business

```python
yelp_business.head(3)
```

```python
# Remove the specified columns
columns_to_drop = ['address', 'city', 'postal_code', 'is_open', 'attributes', 'hours']
yelp_business = yelp_business.drop(columns=columns_to_drop)

# Rename the columns
column_mapping = {
    'business_id': 'id',
    'review_count': 'num_of_reviews',
    'stars': 'avg_rating',
    'categories': 'category'
}
yelp_business = yelp_business.rename(columns=column_mapping)

yelp_business.head(3)
```

```python
# yelp_business['avg_rating'] = yelp_business['avg_rating'].str.replace(',', '.').astype(float)
yelp_business['latitude'] = yelp_business['latitude'].str.replace(',', '.').astype(float)
yelp_business['longitude'] = yelp_business['longitude'].str.replace(',', '.').astype(float)

# Concatenate the DataFrames
combined_df = pd.concat([gmap_metadata, yelp_business])

# Reset the index of the combined DataFrame
beauty_business = combined_df.reset_index(drop=True)
beauty_business.head()
```
